## Indian Premier League (IPL) Regression Analysis

```python
# As usual, we begin by loading the packages we will need

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
# Now we load the data

IPL=pd.read_excel("IPL (assignment) data.xlsx")
IPL.describe()
```

```python
IPL.info()
```

```python
Sumsal = IPL.groupby(['year'])['salaries'] \
                     .sum() \
                     .reset_index() \
                     .rename(columns={'salaries':'allsal'})
Sumsal 
```

```python
IPL = pd.merge(IPL, Sumsal, on = ['year'], how = 'left')
IPL
```

```python
IPL['relsal'] = IPL.salaries / IPL.allsal
IPL['wpc'] = IPL.won / (IPL.played - IPL.noresult)

IPL.head()
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.style.use('seaborn-dark')

sns.regplot(x='relsal', y='wpc', data = IPL, ci = False)
```

```python
IPL['wpc_lag'] = IPL.groupby('team')['wpc'].shift(1)
IPL
```

```python
wpcsal1_lm = smf.ols(formula = 'wpc ~ relsal',
                     data = IPL).fit()
print(wpcsal1_lm.summary())
```

```python
wpcsal2_lm = smf.ols(formula = 'wpc ~ relsal + wpc_lag',
                     data = IPL).fit()
print(wpcsal2_lm.summary())
```

```python
wpcsal3_lm = smf.ols(formula = 'wpc ~ relsal + wpc_lag + C(team)',
                     data = IPL).fit()
print(wpcsal3_lm.summary())
```
