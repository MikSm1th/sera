# I. Building the Dataset: Player Stats

```python
# Read in salary data retrieved from Lahman's Database
Salary = pd.read_csv("Salaries.csv")

# Read in batting
Batting = pd.read_csv("Batting.csv")
```

Load the “Salaries” file, drop any missing values, create a variable for the natural log (ln) of player salary, and rename the column “yearID” as “SalYear”.

```python
#drop missing values

Salary = Salary[Salary.salary >0]
Salary
```

```python
Salary['lnSal'] = np.log(Salary.salary)
Salary
```

```python
Salary = Salary.rename(columns = {'yearID':'SalYear'})
Salary
```

```python
Master = Salary
Master
```

```python
# Load the batting data and sum across stints
Batting = pd.read_csv('Batting.csv')
pd.set_option('display.max_columns', 50)
display(Batting)
```

```python
#Stints

Batting = Batting.groupby(['playerID', 'yearID']).sum()
Batting.reset_index(inplace=True)
Batting
```

```python
Batting = Batting[(Batting.yearID >= 1998) & \
                  (Batting.yearID <= 2006) & \
                  (Batting.AB >= 130)].copy()
Batting
```

```python
Batting['PA'] = Batting.AB + \
                Batting.BB + \
                Batting.HBP + \
                Batting.SH + \
                Batting.SF
```

```python
Batting['OBP'] = (Batting.H + 
                  Batting.BB + 
                  Batting.HBP) / (Batting.AB + 
                                 Batting.BB + 
                                 Batting.HBP + 
                                 Batting.SF)
```

```python
Batting['SLG'] = ((Batting.H - 
                   Batting.Doubles - 
                   Batting.Triples -  
                   Batting.HR) + 
                  2 * Batting.Doubles + 
                  3 * Batting.Triples + 
                  4 * Batting.HR) / Batting.AB
```

```python
Batting['AVG'] = Batting.H / Batting.AB
```

```python
Batting
```

```python
Batting['SalYear'] = Batting.yearID + 1
Batting
```

```python
Master = pd.merge(Batting, Master, on=['SalYear','playerID'])
Master
```

# II. Building the Dataset: Player Info

```python
# Read in People data retrieved from Lahman's Database
People = pd.read_csv("People.csv")
```

```python
Debut = People[['playerID','debut']].copy()
Debut.head()
```

```python
Debut['debutyr'] = Debut.debut.astype(str).str[:4]
Debut.head()
```

```python
Debut = Debut[['playerID','debutyr']]
Debut.head()
```

```python
Master = pd.merge(Master, Debut, 
                  on = ['playerID'],  how = 'left')
Master.head()
```

```python
Master['Exp'] = Master.yearID - \
                Master.debutyr.astype(int)
Master.head(3)
```

```python
Master['Arb'] = np.where((Master.Exp <= 6) &
                         (Master.Exp >= 3), 1,0)
Master['Free'] = np.where(Master.Exp > 6, 1, 0)
Master.head(3)
```

```python
Appearances = pd.read_csv("Appearances.csv")
Appearances.head(3)
```

```python
Appearances = Appearances.groupby(['playerID', 'yearID'])[['G_c','G_1b',\
                                                          'G_2b', 'G_3b', \
                                                          'G_ss', 'G_of', \
                                                          'G_dh']].sum()
Appearances.reset_index(inplace=True)
Appearances.head(3)
```

```python
Appearances['Max_G'] = Appearances[['G_c', 'G_1b', 'G_2b', \
                                    'G_3b', 'G_ss', 'G_of', \
                                    'G_dh']].max(axis=1)
Appearances.head(3)
```

```python
pd.set_option('display.max_rows', 50)
def Position(df):
    if (df['Max_G'] == df['G_c']): return 'C'
    elif (df['Max_G'] == df['G_1b']): return '1B'
    elif (df['Max_G'] == df['G_2b']): return '2B'
    elif (df['Max_G'] == df['G_3b']): return '3B'
    elif (df['Max_G'] == df['G_ss']): return 'SS'
    elif (df['Max_G'] == df['G_of']): return 'OF'
    elif (df['Max_G'] == df['G_dh']): return 'DH'

Appearances['POS'] = Appearances.apply(Position, axis=1)
display(Appearances)
```

```python
Appearances = Appearances[Appearances.Max_G > 0]
Appearances = Appearances[['playerID', 'yearID', 'POS']]
Appearances
```

```python
Appearances['Catch'] = np.where(Appearances.POS == 'C', 1, 0)
Appearances['Infld'] = np.where((Appearances.POS == '2B') | 
                                (Appearances.POS == '3B') |
                                (Appearances.POS == 'SS'), 1, 0)
Appearances
```

```python
Master = pd.merge(Master, Appearances, on=['playerID', 'yearID'], how = 'left')
Master 
```

```python
pd.options.display.float_format = '{:.2f}'.format
MeanSal = Master.groupby(['SalYear','POS'])['salary'].mean().reset_index()
Sal99 = MeanSal[MeanSal.SalYear == 1999]
print(Sal99)
Sal04 = MeanSal[MeanSal.SalYear == 2004]
print(Sal04)
```

# III. Running Regressions

```python
MB_Data_pre = Master[(Master.SalYear >= 1999) & (Master.SalYear <= 2003)]
```

```python
print(Master.columns.tolist())
```

```python
import statsmodels.formula.api as smf
```

```python
reg_form = 'lnSal ~ OBP + SLG + AVG + PA +Arb + Free + Catch + Infld' 
Val_All_lm_pre = smf.ols(formula = reg_form,
                         data = MB_Data_pre).fit()
Val_All_lm_pre.summary()
```

```python
MB_Data_MB = Master[(Master.SalYear >= 2004) & (Master.SalYear <= 2006)]
Val_All_lm_MB = smf.ols(formula = reg_form,
                         data = MB_Data_MB).fit()
Val_All_lm_MB.summary()
```

```python
years = Master.SalYear.unique()
df_list = ['MB_Data_'+str(yr) for yr in years]
reg_list = ['Val_'+str(yr)+'_lm' for yr in years]
df_dict = dict(zip(years, df_list))
reg_dict = dict(zip(years, reg_list))
df_dict
reg_dict
```

```python
regression_list = []
for yr in years:
    df_dict[yr] = Master[(Master.SalYear == yr)]
    reg_dict[yr] = smf.ols(formula=reg_form,
                           data=df_dict[yr]).fit()
    regression_list.append(reg_dict[yr])
```

```python
regression_list[-1].summary()
```

```python
from statsmodels.iolib.summary2 import summary_col
Header = list(years[:4])
regressor_list = ['OBP','SLG','PA','AVG','Arb','Free','Catch','Infld','Intercept']
Table_Pre = summary_col(regression_list[:4],
                         regressor_order=regression_list, 
                         stars=True,
                         float_format="'%.3f'",
                         model_names = Header)
print(Table_Pre)
```

```python
Header = list(years[4:])
regressor_list = ['OBP','SLG','PA','AVG','Arb','Free','Catch','Infld','Intercept']
Table_MB = summary_col(regression_list[4:],
                         regressor_order=regression_list, 
                         stars=True,
                         float_format="'%.3f'",
                         model_names = Header)
print(Table_MB)
```
